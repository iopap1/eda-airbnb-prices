Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the dataset

```python
# Using seaborn's built-in dataset (about restaurant bills)
df = sns.load_dataset("tips")

# Show first 5 rows
df.head()
```

Explore the data

```python
# Dataset shape
print("Shape:", df.shape)

# Summary statistics
df.describe()
```

First visualization

```python
sns.barplot(data=df, x="day", y="total_bill")
plt.title("Average Bill by Day")
plt.show()
```
